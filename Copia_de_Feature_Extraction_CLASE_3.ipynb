```python
import cv2
from google.colab.patches import cv2_imshow
```

```python
cv2.__version__
```

```python
input_image = cv2.imread("/content/drive/MyDrive/Colab Notebooks/Robot.jpg")
cv2_imshow(input_image)
```

```python
input_image = cv2.resize(input_image, (500, 350))
cv2_imshow(input_image)
```

```python
input_image.shape
```

```python
gray_image = cv2.cvtColor(input_image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```python
print(gray_image)
print(gray_image.shape)
```

## SIFT

```python
sift = cv2.xfeatures2d.SIFT_create()
```

```python
kp_sift, des_sift = sift.detectAndCompute(gray_image, None)
```

```python
print("Keypoints: ", len(kp_sift))
print("Descriptores: ", des_sift.shape[1])
```

```python
img_sift = cv2.drawKeypoints(gray_image,kp_sift,input_image,flags=cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
cv2_imshow(img_sift)
```

## SURF

```python
surf = cv2.xfeatures2d.SURF_create(5000)
```

```python
kp_surf, des_surf = surf.detectAndCompute(gray_image,None)
```

```python
print("Keypoints: ", len(kp_surf))
print("Descriptores: ", des_surf.shape[1])
```

```python
img_surf = cv2.drawKeypoints(gray_image,kp_surf,None,(255,0,0),4)
cv2_imshow(img_surf)
```

## ORB

```python
orb = cv2.ORB_create(5000)
```

```python
kp_orb, des_orb = orb.detectAndCompute(gray_image, None)
```

```python
print("Keypoints: ", len(kp_orb))
print("Descriptores: ", des_orb.shape[1])
```

```python
img_orb = cv2.drawKeypoints(gray_image, kp_orb, input_image, (0,255,0))
cv2_imshow(img_orb)
```
